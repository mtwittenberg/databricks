#### functions

```python
def initialize_driver_short():
    ## Incognito driver
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--incognito")
    
    ## Default download bucket
    #     prefs = {'download.default_directory': location}
    #     chrome_options.add_experimental_option('prefs', prefs)

    ## Running driver
    driver = webdriver.Chrome(options=chrome_options)
    
    return driver
```

```python
def landing_page(browser):
    url = 'http://www.njrcentre.org.uk/njrcentre/Healthcare-providers/Accessing-the-data/StatsOnline/NJR-StatsOnline'
    download_page = browser.get(url)
    browser.implicitly_wait(2) 
```

```python
def get_text(browser):
    path = '//*[@id="dnn_ctr689_ViewNjrStatsOnline_Title1_pnlNationalText"]'
    element_ = driver.find_element_by_xpath(path).text
    return(element_)
```

```python
def table_parser(browser):
    
    ## Wait for visability
    WebDriverWait(driver, 60, 2).until(ec.visibility_of_element_located([By.ID, 'dnn_ctr689_ViewNjrStatsOnline_SumaryTable1_NhsTable']))
    
    ## Get main table
    table = browser.find_element_by_id('dnn_ctr689_ViewNjrStatsOnline_SumaryTable1_NhsTable')
    length_rows = len(table.find_elements_by_tag_name("tr"))
    
    ## Final Loop from table
    total_list = []
    for r in range(2, length_rows + 1): 
        xpath_row = f'//*[@id="dnn_ctr689_ViewNjrStatsOnline_SumaryTable1_NhsTable"]/tbody/tr[{r}]'
        get_row = table.find_element_by_xpath(xpath_row)
        elements_row = get_row.find_elements_by_tag_name('td')

        ## Append row items
        temp_list = []
        for i in elements_row:
            item = i.text
            temp_list.append(item)
        total_list.append(temp_list)
        
    return(total_list)
```

#### RUN

```python
driver = initialize_driver_short()
landing_page(driver)

text = get_text(driver)
raw = table_parser(driver)

# time.sleep(15)
driver.close()
```
